```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset
# Assuming your dataset is in a CSV file named 'data.csv'
df = pd.read_csv('datawlabel.csv')

# Choose the feature you want to visualize
# Replace 'feature_name' with the actual name of the feature/column in your dataset
feature_name = 'closnesscentrality'

# Create a basic plot
plt.figure(figsize=(8, 6))
plt.hist(df[feature_name], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Feature Values')
plt.ylabel('Frequency')
plt.title(f'Distribution of {feature_name}')
plt.grid(True)
plt.show()
```
